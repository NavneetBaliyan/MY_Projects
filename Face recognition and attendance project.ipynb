```python
import pandas as pd
import numpy as np
import cmake
import dlib
import cv2
import face_recognition
import matplotlib.pyplot as plt
```

```python
imageElon=face_recognition.load_image_file('Elon_Musk_Royal_Society_(crop1).jpg')
imageElon=cv2.cvtColor(imageElon,cv2.COLOR_BGR2RGB)

elontest=face_recognition.load_image_file('elon_musk2.jfif')
elontest=cv2.cvtColor(elontest,cv2.COLOR_BGR2RGB)

faceloc=face_recognition.face_locations(imageElon)[0]
encodeElon=face_recognition.face_encodings(imageElon)[0]
cv2.rectangle(imageElon,(faceloc[3],faceloc[0]),(faceloc[1],faceloc[2]),(255,0,255),2)

faceloctest=face_recognition.face_locations(elontest)[0]
encodeElontest=face_recognition.face_encodings(elontest)[0]
cv2.rectangle(elontest,(faceloctest[3],faceloctest[0]),(faceloctest[1],faceloctest[2]),(255,0,255),2)


#comparing these faces
results=face_recognition.compare_faces([encodeElon],encodeElontest)
facedist=face_recognition.face_distance([encodeElon],encodeElontest)
print(results,facedist)
cv2.putText(elontest,f'{results}{round(facedist[0],2)}',(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)

cv2.imshow('Elon Musk',imageElon)
cv2.imshow('Elon test',elontest)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imageNav=face_recognition.load_image_file('navneet_pic.jfif')
imageNav=cv2.cvtColor(imageNav,cv2.COLOR_BGR2RGB)

Navtest=face_recognition.load_image_file('030.JPG')
IMG_SIZE=720
Navtest=cv2.resize(Navtest,(IMG_SIZE,IMG_SIZE))
Navtest=cv2.cvtColor(Navtest,cv2.COLOR_BGR2RGB)

faceloc=face_recognition.face_locations(imageNav)[0]
encodeNav=face_recognition.face_encodings(imageNav)[0]
cv2.rectangle(imageNav,(faceloc[3],faceloc[0]),(faceloc[1],faceloc[2]),(255,0,255),2)

faceloctest=face_recognition.face_locations(Navtest)[0]
encodeNavtest=face_recognition.face_encodings(Navtest)[0]
cv2.rectangle(Navtest,(faceloctest[3],faceloctest[0]),(faceloctest[1],faceloctest[2]),(255,0,255),2)


#comparing these faces
results=face_recognition.compare_faces([encodeNav],encodeNavtest)
facedist=face_recognition.face_distance([encodeNav],encodeNavtest)
print(results,facedist)
cv2.putText(Navtest,f'{results}{round(facedist[0],2)}',(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)

cv2.imshow('Navneet',imageNav)
cv2.imshow('Navneet test',Navtest)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
